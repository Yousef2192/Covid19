```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
df
```

```python
plt.figure(figsize=(20,6))

#plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.plot(df_Turkey.index, df_Turkey['total_cases'])

plt.title('Evolution of Confirmed Covid-19 cases over time in the word', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Confirmed cases', fontsize=1)
plt.show()
```

```python
# Using line-plot
plt.figure(figsize=(20,6))
#plt.plot(df_Turkey, df_Turkey['total_cases'], label='Confirmed')
plt.plot(df_Turkey.index, df_Turkey['total_deaths'], label='Total Deaths')
plt.plot(df_Turkey.index, df_Turkey['new_cases'], label='New')
plt.plot(df_Turkey.index, df_Turkey['new_deaths'], label='New Deaths')
plt.legend(loc=0)
plt.show()
```
